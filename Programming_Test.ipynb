```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# El documento con los datos de todos los municipios viene en dos partes, así que lo fucionamos para que
# se junten los datos de los estados del 1 al 15 y del 16 al 32

df1 = pd.read_csv('IML_2020_1.csv') # Estados del 1 al 15
df2 = pd.read_csv('IML_2020_2.csv') # Estados del 16 al 32
#CVE_LOC	ENT	NOM_ENT	MUN	NOM_MUN	LOC	NOM_LOC	POB_TOT	ANALF	SBASC	OVSDE	OVSEE	OVSAE	OVPT	OVHAC	OVSREF	IM_2020	GM_2020	IMN_2020

# Concatenate the dataframes vertically
df = pd.concat([df1, df2])

# Reset the index of the joined dataframe
df = df.reset_index(drop=True)
df.to_csv('IML_2020_FULL.csv', index=True)
```

```python
# Mostramos la descripción del DataFrame
print(df.describe())
df
```

```python
# Crear un diccionario con los alias deseados

dfdicc = pd.DataFrame({'Variable':['CVE_LOC','ENT', 'NOM_ENT','MUN','NOM_MUN','LOC','NOM_LOC','POB_TOT','ANALF','SBASC','OVSDE','OVSEE','OVSAE','OVPT','OVHAC','OVSREF','IM_2020','GM_2020','IMN_2020'],
'Descripcion':[
'Clave geográfica',
'Clave de la entidad federativa',
'Nombre de la entidad federativa',
'Clave del municipio',
'Nombre del municipio',
'Clave de la localidad',
'Nombre de la localidad',
'Población total',
'Porcentaje de población de 15 años o más analfabeta',
'Porcentaje de población de 15 años o más sin educación básica',
'Porcentaje de ocupantes en viviendas particulares habitadas sin drenaje ni excusado',
'Porcentaje de ocupantes en viviendas particulares habitadas sin energía eléctrica',
'Porcentaje de ocupantes en viviendas particulares habitadas sin agua entubada',
'Porcentaje de ocupantes en viviendas particulares habitadas con piso de tierra',
'Porcentaje de ocupantes en viviendas particulares habitadas con hacinamiento',
'Porcentaje de ocupantes en viviendas particulares habitadas sin refrigerador',
'Índice de marginación a nivel localidad, 2020',
'Grado de marginación a nivel localidad, 2020',
'Índice de marginación normalizado a nivel localidad, 2020']
})


print(dfdicc)

```

```python

# Obtener el índice del valor máximo de la columna 'IMN_2020'
indice_max_marginal = df['IMN_2020'].idxmin()

# Obtener la ciudad correspondiente al valor máximo de 'IMN_2020'
ciudad_max_marginal = df.loc[indice_max_marginal, 'ENT']

# Usé 
# ´´´
df['POB_TOT'] = df['POB_TOT'].str.replace(',', '')
df['POB_TOT'] = df['POB_TOT'].apply(int)
# ´´´
# para corregir el problema de las comas separando cada 3 digitos de los números
# Mostramos algunos hallazgos interesantes

print("Media de la columna 'Población':", (df['POB_TOT'].mean()))

first_value = df['POB_TOT'].iloc[0]
data_type = type(first_value)

print("Primer valor en la columna 'POB_TOT':", first_value)
print("Tipo de dato del primer valor:", data_type)
```

```python
first_value_GM = df['GM_2020'].iloc[0]
data_type_GM = type(first_value_GM)

print("Primer valor en la columna 'gm_2020':", first_value_GM)
print("Tipo de dato del primer valor:", data_type_GM)


count = df['GM_2020'].value_counts()
print(count)
```

```python
#Agrupar los datos por la columna "NOM_ENT" y calcular el porcentaje de cada valor dentro de cada estado
df_to_graph = df.groupby('NOM_ENT')['GM_2020'].value_counts(normalize=True) * 100

#Pivotear los datos para tener los valores como columnas y los estados como índice
df_to_graph = df_to_graph.unstack()

# Definir el nuevo orden de las columnas
orden = ['Muy bajo', 'Bajo', 'Medio', 'Alto', 'Muy alto']

# Reordenar las columnas en el DataFrame
df_to_graph = df_to_graph.reindex(columns=orden)

df_to_graph
```

```python
#Para que la gráfica se pueda apreciar mejor
fig, ax = plt.subplots(figsize=(10, 6))

df_to_graph.plot.bar(ax=ax)
plt.xlabel('Estado')
plt.ylabel('Porcentaje')
plt.title('Distribución de los grados marginales por estado')
plt.legend(title='Grado Marginal 2020')
plt.savefig('Grado Marginal por Estado.jpg')
```

En general me parece que las regiones de color ROJO y las regiones de color MORADO comprenden un área mucho menor que la combinación de los otrs 3 estados de marginalización.

```python
#Este formato me parece más explicativo para mí, se lo comparto

df_to_graph.plot.bar(stacked=True)
plt.xlabel('Estado')
plt.ylabel('Porcentaje')
plt.title('Distribución de los grados marginales por estado')
plt.legend(title='Grado Marginal 2020')
plt.savefig('Grado Marginal por Estado Barra a 100.jpg')
```

```python

# Filtrar los datos para considerar solo las localidades de menos de 5,000 habitantes
df_filtered = df[df['POB_TOT'] < 5000]

# Calcular el porcentaje de analfabetismo respecto al total de la población
porcentaje_analfabetismo = (df_filtered['ANALF'] / df_filtered['POB_TOT']) * 100

# Calcular el porcentaje de poblaciones en localidades de menos de 5,000 habitantes respecto al total de la población
porcentaje_poblaciones_menos_5000 = (df_filtered['POB_TOT'] / df['POB_TOT'].sum()) * 10000

# Graficar la relación entre los dos porcentajes
plt.scatter(porcentaje_poblaciones_menos_5000, porcentaje_analfabetismo)
plt.xlabel('Porcentaje de analfabetismo')
plt.ylabel('Poblaciones en localidades < 5,000 hab.')
plt.title('Relación entre analfabetismo y poblaciones < 5,000 hab.')
plt.show()

# porcentaje_poblaciones_menos_5000 Lo tenía aquí para vizualizar los valores en una lista
```

Se puede apreciar el crecimiento del analfabetismo entre menos población existe en un poblado. Ésto quizá por el incremento en el apoyp a las grandes comunidades.
